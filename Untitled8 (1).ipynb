```python
import pandas as pd
import numpy as np

# Read the CSV file directly into a DataFrame
df = pd.read_csv('D:\DATOS\DESCARGAS\IRLTLT01DKQ156N.csv')

# Convert 'DATE' column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Set 'DATE' column as index
df.set_index('DATE', inplace=True)

# Apply logarithmic transformation
df['Log_IRLTLT01DKQ156N'] = np.log(data[data['IRLTLT01DKQ156N'] > 0]['IRLTLT01DKQ156N'])
df['Log_IRLTLT01DKQ156N'].fillna(0, inplace=True)

# Resample to quarterly intervals
df_quarterly = df.resample('Q').mean()

# Reset index to make 'DATE' a column again
df_quarterly = df_quarterly.reset_index()

# Print the aggregated dataset
print(df_quarterly)


```
